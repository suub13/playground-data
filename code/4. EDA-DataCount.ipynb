#### skills 데이터 셋 확인하기

```python

def getskillsSetList():
    with open( "../data/crawled_data_combined.csv", encoding="utf-8") as f:
        reader = csv.reader(f)
        # 태그 하나씩 쪼개서 넣을 리스트 (set으로 중복 제거 전)
        skillsList = []

        # csv 파일 리더로, 한 row씩 읽기
        for row in reader:

            # row list의 10번째 element인 tags string을 ''안 단어들로 나눠서 list만들기
            skills = re.findall(r"'(.*?)'", row[10])     
            # print(row[10])

            # 각 줄의 태그를 tagList에 넣는다.
            skillsList += skills

            # count += 1
        
        skillsSet = set(skillsList)
        skillsList = list(skillsSet)
        # print(skillsSet)
        # print(len(skillsList))
        # print(len(skillsSet))
        return skillsList
```

```python
skillList = getskillsSetList()
print(len(skillList))
```

####  ListToDict(lst) - skills List로 딕셔너리 만들기

```python
def ListToDict(lst):
    toDict = dict()

    for item in lst:
        toDict[item.upper()] = 0
    
    return toDict
    
```

```python
skillDict = ListToDict(skillList)
len(skillDict)
```

### save csv file

```python
import csv

def saveCSV(dict, title):
    # Specify the CSV file path

    # Open the CSV file in write mode
    with open(title, "w", newline="") as file:
        writer = csv.writer(file)

        # Write the header row with column names
        writer.writerow(["Key", "Value"])

        # Write each key-value pair as a row in the CSV file
        for key, value in dict.items():
            writer.writerow([key, value])

    print("Dictionary keys and values saved to CSV file successfully.")
```

## Qualification 데이터 속 개수 확인

```python
import csv
import ast

qual_dict = skillDict.copy()

with open("./공고별 skill 정리/qual_pref_skill_combined.csv","r",encoding="949") as f:
    reader = csv.reader(f)
    rows = list(reader)

    for row in rows[1:]:
        quals = ast.literal_eval(row[1]) if row[1] else []
        for qual in quals:
            if qual in qual_dict.keys():
                qual_dict[qual] = qual_dict[qual] + 1
                print(qual)
    print(qual_dict)
saveCSV(qual_dict, "./skill 별 빈도수 정리/qual_count.csv")
```

## Preferred 데이터

```python
import csv
import ast

pref_dict = skillDict.copy()

with open("./공고별 skill 정리/qual_pref_skill_combined.csv","r",encoding="949") as f:
    reader = csv.reader(f)
    rows = list(reader)

    for row in rows[1:]:
        prefs = ast.literal_eval(row[2]) if row[2] else []
        for pref in prefs:
            if pref in qual_dict.keys():
                pref_dict[pref] = pref_dict[pref] + 1
                print(pref)
    print(pref_dict)
saveCSV(pref_dict, "./skill 별 빈도수_정리 전/pref_count.csv")
```

## skill 데이터

```python
import csv
import ast

skill_dict = skillDict.copy()

with open("./공고별 skill 정리/qual_pref_skill_combined.csv","r",encoding="949") as f:
    reader = csv.reader(f)
    rows = list(reader)

    for row in rows[1:]:
        row_upper = row[3].upper()
        skills = ast.literal_eval(row_upper) if row_upper else []
        for skill in skills:
            if skill in qual_dict.keys():
                skill_dict[skill] = skill_dict[skill] + 1
                print(skill)
    print(skill_dict)
saveCSV(skill_dict, "./skill 별 빈도수_정리 전/skill_count.csv")
```

## skill 명 같은지 확인 후, 한 데이터 프레임 안에 qual, pref, skill 빈도 테이블 csv로 저장하기

```python
qual_key = list(qual_dict.keys())
qual_value = list(qual_dict.values())

pref_key = list(pref_dict.keys())
pref_value = list(pref_dict.values())

skill_key = list(skill_dict.keys())
skill_value = list(skill_dict.values())
len(skill_value)
```

#### skill 명 같은지 확인 아니면 해당 열 번호 print

```python
rate = []
for row in range(len(qual_key)):
    if (qual_key[row] == pref_key[row] == skill_key[row]):
        calcuated = (int(qual_value[row]) * 3) + pref_value[row] + skill_value[row]
        rate.append(calcuated)
print(rate)
len(rate)
```

#### skill 명 + quals + prefs + skills 저장

```python
import pandas as pd

# Create the DataFrame
df = pd.DataFrame({'name': qual_key, 'qualification': qual_value, 'preferred': pref_value, 'skills': skill_value, 'qual*3+pref+skill':rate})

# Save the DataFrame as a CSV file with column names
df.to_csv('./skill 별 빈도수_정리 전/qual_pref_skill_count.csv', index=False, encoding="utf-8")
```
